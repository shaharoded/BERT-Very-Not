```python
import pickle
from collections import defaultdict

# Replace with the actual path
file_path = 'snli_mnli_llm_validation.pkl'
with open(file_path, 'rb') as f:
    data = pickle.load(f)

# 2. Group by premise
by_premise = defaultdict(list)
for ex in data:
    by_premise[ex['premise']].append(ex)

# 3. Find the first two premises that have both an entailment and a contradiction
pairs = []
for premise, exs in by_premise.items():
    ents = [e for e in exs if e['label'] == 0]   # entailments
    cons = [e for e in exs if e['label'] == 2]   # contradictions
    if ents and cons:
        pairs.append((ents[0], cons[0]))
    if len(pairs) == 50:
        break

# 4. Print them
for idx, (orig, aug) in enumerate(pairs, 1):
    print(f"\nPair {idx}:")
    print("Original (Entailment):")
    print(f"  Premise:    {orig['premise']}")
    print(f"  Hypothesis: {orig['hypothesis']}")
    print(f"  Label:      {orig['label']}")
    print("\nAugmented (Contradiction):")
    print(f"  Premise:    {aug['premise']}")
    print(f"  Hypothesis: {aug['hypothesis']}")
    print(f"  Label:      {aug['label']}")
```

```python
import pickle
from collections import defaultdict

# Load data
file_path = 'snli_mnli_squad_spacy_validation.pkl'
with open(file_path, 'rb') as f:
    data = pickle.load(f)

# Step 1: Index all entailments by premise
entailments = defaultdict(list)
for ex in data:
    if ex['label'] == 0:
        entailments[ex['premise']].append(ex)

# Step 2: Find contradictions from SQuAD-style augmentation
spacy_contradictions = []
for ex in data:
    if ex['label'] == 2:
        # Only keep if there’s a matching entailment for the same premise
        if ex['premise'] in entailments:
            spacy_contradictions.append((entailments[ex['premise']][0], ex))

# Step 3: Take the last 50
print(f"Found {len(spacy_contradictions)} contradiction pairs with matched entailments.")
last_50 = spacy_contradictions[-50:]

# Step 4: Print the results
for idx, (ent, contra) in enumerate(last_50, start=1):
    print(f"\nPair {idx}:")
    print("Original (Entailment):")
    print(f"  Premise:    {ent['premise']}")
    print(f"  Hypothesis: {ent['hypothesis']}")
    print(f"  Label:      {ent['label']}")
    print("Augmented (Contradiction):")
    print(f"  Premise:    {contra['premise']}")
    print(f"  Hypothesis: {contra['hypothesis']}")
    print(f"  Label:      {contra['label']}")
```
